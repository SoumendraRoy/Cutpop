```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import corner
```

# Result for Truth

```python
with h5py.File("hyperparameter_true.h5",'r') as hf:
    f1 = np.array(hf['f1'])
    mu1 = np.array(hf['mu1'])
    sigma1 = np.array(hf['sigma1'])
    mu2 = np.array(hf['mu2'])
    sigma2 = np.array(hf['sigma2'])
    a = np.array(hf['a'])
    b = np.array(hf['b'])
hf.close()
```

```python
import corner
figure = corner.corner(np.array([f1,mu1,sigma1,mu2,sigma2,a,b]).T,
                       labels=[r'$f_1$',r'$\mu_1$',r'$\sigma_1$',r'$\mu_2$', r'$\sigma_2$',r'$a$',r'$b$'])
true = np.array([0.7,35.0,3.0,55.0,1.0,3.0,9.0])
# Extract the axes
axes = np.array(figure.axes).reshape((len(true), len(true)))

# Loop over the diagonal
for i in range(len(true)):
    ax = axes[i, i]
    ax.axvline(true[i], color="r")



# Loop over the histograms
for yi in range(len(true)):
    for xi in range(yi):
        ax = axes[yi, xi]
        ax.axvline(true[xi], color="r")
        ax.axhline(true[yi], color="r")
        ax.plot(true[xi], true[yi], "sr")
```
